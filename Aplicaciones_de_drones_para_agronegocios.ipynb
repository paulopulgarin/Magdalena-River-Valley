**Numero de Dimensiones**

NumPy Arrays proporciona el atributo `ndim` que devuelve un número entero que nos dice cuántas dimensiones tiene la matriz.

````
a = np.array(42)
b = np.array([1, 2, 3, 4, 5])
c = np.array([[1, 2, 3], [4, 5, 6]])
d = np.array([[[1, 2, 3], [4, 5, 6]], [[1, 2, 3], [4, 5, 6]]])

print('Dimensiones de arreglo "a":', a.ndim, 'dimensiones')
print('Dimensiones de arreglo "b":', b.ndim, 'dimensiones')
print('Dimensiones de arreglo "c":', c.ndim, 'dimensiones')
print('Dimensiones de arreglo "d":', d.ndim, 'dimensiones')
````

**Arreglos de mas dimensiones**

Un *array* puede tener cualquier cantidad de dimensiones.


````
arr5d = np.array([1, 2, 3, 4], ndmin=5)

print(arr5d)
print('Dimensiones de arreglo "arr5d"', arr5d.ndim)
````

Como podemos este resultado?

En este arreglo (`arr5d`) la dimensión más interna (5º dim) tiene 4 elementos, el 4º dim tiene 1 elemento que es el vector, el 3º dim tiene 1 elemento que es la matriz con el vector, el 2º dim tiene 1 elemento que es una matriz 3D y 1st dim tiene 1 elemento que es una matriz 4D.

**NOTA**: NumPy tiene un submódulo completo dedicado a operaciones matriciales llamado `numpy.mat`.

#### Linea
Generalmente, se utiliza un gráfico de líneas para presentar las observaciones recopiladas a intervalos regulares.

El eje x representa el intervalo regular, como el tiempo. El eje y muestra las observaciones, ordenadas por el eje x y conectadas por una línea.

Se puede crear una gráfica de línea llamando a la función ```plot()``` y pasando los datos del eje x para el intervalo regular y el eje y para las observaciones.

```python

```

Line plots are useful for presenting time series data as well as any sequence data where there is an ordering between observations.

The example below creates a sequence of 100 floating point values as the x-axis and a sine wave as a function of the x-axis as the observations on the y-axis. The results are plotted as a line plot.

````
x = [x*0.1 for x in range(100)]
y = sin(x)
````

````
x1 = [x*0.1 for x in range(100)]
x2 = [x*0.5 for x in range(100)]
y1 = sin(x1)
y2 = sin(x2)

plt.plot(x1, y1, color='black', marker='o', label = 'Linea Negra')
plt.plot(x2, y2, color='red', marker='*', label = 'Linea Roja')
plt.xlabel('Eje X', fontsize = 16)
plt.ylabel('Eje Y', fontsize = 16)
plt.title('Mi primer plot en Python', fontsize = 20)
plt.grid()
plt.legend()
plt.show()
````

#### Histograma

Un diagrama de histograma se usa generalmente para resumir la distribución de una muestra de datos.

El eje x representa intervalos o intervalos discretos para las observaciones. Por ejemplo, las observaciones con valores entre 1 y 10 pueden dividirse en cinco intervalos, los valores [1,2] se asignarían al primer intervalo, [3,4] se asignarían al segundo intervalo, y así sucesivamente.

El eje y representa la frecuencia o recuento del número de observaciones en el conjunto de datos que pertenecen a cada contenedor.

Esencialmente, una muestra de datos se transforma en un gráfico de barras donde cada categoría en el eje x representa un intervalo de valores de observación.

Se puede crear un diagrama de histograma llamando a la función ```hist()``` y pasando una lista o matriz que represente la muestra de datos.

```python
# create histogram plot
plt.hist(x)
```

Los histogramas son valiosos para resumir la distribución de muestras de datos.

El siguiente ejemplo crea un conjunto de datos de 1000 números aleatorios extraídos de una distribución gaussiana estándar y luego traza el conjunto de datos como un histograma.

````
# import module
from numpy.random import randn

# random numbers drawn from a Gaussian distribution
x = randn(1000)
````

Crear histograma:

````
# create histogram plot
plt.hist(x)
plt.grid()
plt.title('Histograma')
plt.show()
````

Ejecutando el ejemplo, podemos ver que la forma de las barras muestra la curva en forma de campana de la distribución gaussiana. Podemos ver que la función eligió automáticamente el número de bins, en este caso dividiendo los valores en grupos por valor entero.

A menudo, la elección cuidadosa del número de contenedores puede ayudar a exponer mejor la forma de la distribución de datos. El número de bins se puede especificar configurando el argumento *bins*; por ejemplo:

````
# create histogram plot
plt.hist(x, bins=100)
````

````
plt.hist(df['electricity'])
````

#### Scatter (Puntos)

Un diagrama de dispersión (o "diagrama de dispersión") se usa generalmente para resumir la relación entre dos muestras de datos emparejadas.

Muestras de datos emparejados significa que se registraron dos medidas para una observación determinada, como el peso y la altura de una persona.

El eje x representa los valores de observación para la primera muestra y el eje y representa los valores de observación para la segunda muestra. Cada punto del gráfico representa una única observación.

Los diagramas de dispersión se pueden crear llamando a la función ```scatter()``` y pasando las dos matrices de muestra de datos.

```python
plt.scatter (x, y)
```

Los gráficos de dispersión son útiles para mostrar la asociación o correlación entre dos variables. Se puede cuantificar una correlación, como una línea de mejor ajuste, que también se puede dibujar como un gráfico de líneas en el mismo gráfico, lo que hace que la relación sea más clara.

Un conjunto de datos puede tener más de dos medidas (variables o columnas) para una observación determinada. Una matriz de gráficos de dispersión es un carro que contiene gráficos de dispersión para cada par de variables en un conjunto de datos con más de dos variables.

El siguiente ejemplo crea dos muestras de datos que están relacionadas. El primero es una muestra de números aleatorios extraídos de un gaussiano estándar. El segundo depende del primero al agregar un segundo valor gaussiano aleatorio al valor de la primera medida.
